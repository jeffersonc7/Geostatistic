### __Multiple Linear Regression Model__
we considering the theoretical linear regression model : 

let $\mu$ be the expected value of the response variable
$$ \mu = E(\mathbf{Y| X_1,X_2,\dots,X_n}) \qquad \equiv \qquad \mu = {\beta}_0 + {\beta}_1 {X}_1 + {\beta}_2 {X}_2 +\dots + \beta_n {X}_n$$

now we define the Linear regression model in matrix form 
$$ \mathbf{Y} = \mu + \epsilon \qquad \equiv \qquad \mathbf{Y} = \mathbf{X}\boldsymbol{\beta} + {\varepsilon}$$
where : 

- $Y$ : Vector of observations 
- $X$ : Design matrix (explanatory variables)
- $\beta$ : Vector of regression coeficients $({\beta}_0, {\beta}_1, \dots, \beta_n )$
- $\epsilon \sim \mathcal{N}(0,\sigma^2)$ : Error term

after observing the data, we obtain : 
$$\hat{Y} = \hat{\beta}_0 + \hat{\beta}_1 {X}_1 + \hat{\beta}_2 {X}_2 +\dots + \hat\beta_n {X}_n$$
To estimate the regression coeficients, we apply the Ordinary Least Squares __(OLS)__:

$$ \hat{\beta}= (\mathbf{X}^T \mathbf{X})^{-1}  \mathbf{X}^T \mathbf{Y} $$

```python
# INPUT DATA 
import numpy as np
import pandas as pd

np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 100 muestras de una variable independiente
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3*X + ruido
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Agregar una constante a X (para el intercepto)
X = sm.add_constant(X)

# Crear el modelo de regresión lineal
model = sm.OLS(y, X)

# Ajustar el modelo
results = model.fit()

# Ver el resumen del modelo
print(results.summary())

# Hacer predicciones
y_pred = results.predict(X)

# Calcular el MSE
mse = mean_squared_error(y, y_pred)
print(f"Error cuadrático medio (MSE): {mse}")

# Graficar los resultados
plt.scatter(X[:, 1], y, color='blue', label='Datos reales')  # X[:, 1] es la variable independiente sin la constante
plt.plot(X[:, 1], y_pred, color='red', linewidth=2, label='Línea de regresión')
plt.xlabel('Variable independiente X')
plt.ylabel('Variable dependiente y')
plt.title('Regresión Lineal con statsmodels')
plt.legend()
plt.show()
```
