### __Libraries__

```python
# Required Dependencies 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

### __Data Types__

```python
# Import Data
data = pd.read_csv('NY-House-Dataset/NY-House-Dataset.csv')
# data types 
print(data.dtypes)
display(data.head(5))
display(data.tail(4))
```

```python
# DESCRIBE DATA (float - variables)
display(data.describe())
display(data.describe(include='object'))
```

```python
data.info()
num_data = data.select_dtypes(include='number')
```

```python
num_data.info()
```

### __Outliers__

```python
import matplotlib.pyplot as plt 

plt.style.use('seaborn-v0_8')
fig,ax = plt.subplots(2,3,figsize=(17,10))

cols = num_data.columns
k = 0 
for i in range(2):
    for j in range(3):

        x = num_data[cols[k]]
        ax[i,j].boxplot(x)

        if (x>0).all():
            ax[i,j].set_yscale('log')
            ax[i,j].set_ylabel("LOG - SCALE")
        else:
            ax[i,j].set_ylabel("SCALE")

        ax[i,j].set_title(f'BOXBLOT - {num_data.columns[k]}',fontweight= 'bold')
        ax[i,j].set_xlabel(f'{num_data.columns[k]}')

        k += 1
plt.tight_layout()
plt.show()
```

### __Data Transform__

```python
# log - Transform
import numpy as np 

num_data['LOG-PRICE'] = np.log(num_data['PRICE'])
num_data['LOG-PROPERTYSQFT'] = np.log(num_data['PROPERTYSQFT'])

# Describe data
num_data[['LOG-PRICE','LOG-PROPERTYSQFT']].describe()
```

```python
import matplotlib.pyplot as plt 

# Box plots in transformed variables 
fig,ax = plt.subplots(1,2,figsize=(15,6))
plt.style.use('seaborn-v0_8')

x = ['LOG-PRICE','LOG-PROPERTYSQFT']

for i in range(2):
    ax[i].boxplot(num_data[x[i]])
    ax[i].set_title(f'BOX PLOT - {x[i]}',fontweight = 'bold')
    ax[i].set_ylabel('Frecuency')
    ax[i].set_xlabel(f'{x[i]}')
plt.show()
```

### __Import Data__

```python
# IMPORT DATA TO CSV
num_data.to_csv('NY-House-Numerical.csv',index=False)
```
