#### __SUMMARY__

```python
# Import data 
import pandas as pd 

# Original Data set 
data = pd.read_csv("NY-House-Dataset/NY-House-Dataset.csv")
# Clean Data (cuantitative variables )
num_data = pd.read_csv("NY-House-Dataset/NY-House-Numerical.csv")

display(data.head(8))
display(data.describe(include='object'))
display(data.head(8))
display(num_data.describe())
print(f"DATA : {data.columns} \n NUM_DATA : {num_data.columns}")
```

#### __DESCRIPTIVE CUANTITATIVE VARIABLES__

```python
# HISTOGRAMS
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(2,3,figsize=(17,10))

cols = ['LOG-PRICE','LOG-PROPERTYSQFT','BEDS','BATH','LATITUDE','LONGITUDE']

k = 0
for i in range(2):
    for j in range(3):
        # Histogram 
        ax[i,j].hist(num_data[cols[k]],color='midnightblue',bins=15,density=True)
        # Density curve 
        sns.kdeplot(num_data[cols[k]], ax=ax[i,j], color='red')

        ax[i,j].set_title(f'HISTOGRAM - {cols[k]}',fontweight = 'bold')
        ax[i,j].set_ylabel('Frequency')
        ax[i,j].set_xlabel(f'{cols[k]}')
        k += 1

plt.tight_layout()
plt.show()
```

```python
# BAR PLOTS 
import matplotlib.pyplot as plt 
import numpy as np

fig, ax = plt.subplots(1,3,figsize=(15,5))
plt.style.use('seaborn-v0_8')

cols1 = ['BEDS', 'BATH','PROPERTYSQFT']

k1 = 0
for i in range(3): 
    if k1 < len(cols1):
            
        # show the 4 first categories
        category_counts = num_data[cols1[k1]].value_counts().head(4)
        # bar colors 
        colors = plt.cm.Blues(np.linspace(1,0.6 , len(category_counts)))
        # barplot
        ax[i].bar(category_counts.index.astype(str), category_counts.values,color=colors)

        ax[i].set_title(f"BAR PLOT - {cols1[k1]}",fontweight= 'bold')
        ax[i].set_xlabel(f'{cols1[k1]}')
        ax[i].set_ylabel('Frequency')
        k1 += 1
plt.tight_layout()
plt.show()
```

```python
# CORRELATION MATRIX

import seaborn as sns 
import matplotlib.pyplot as plt

cols1 = ['LOG-PRICE','LOG-PROPERTYSQFT', 'BEDS', 'BATH' ,'LATITUDE','LONGITUDE']

# Create the matrix
correlation_matrix = num_data[cols1].corr()

# Heatmap 
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix,annot=True,cmap='Blues',fmt='.2f',vmin=-1,vmax=1)
plt.title('CORRELATION MATRIX (HEATMAP)',fontweight='bold')
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.show()
```
